```python
import numpy as np
import pandas as pd
data_train=np.loadtxt(r"C:\Users\dell\Downloads\0000000000002419_training_ccpp_x_y_train.csv",delimiter=",")
df=pd.DataFrame(data_train)
df
```

```python
X_train_data=data_train[:,0:4]
Y_train_data=data_train[:,4]
X_train_data.shape,Y_train_data.shape
```

```python
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
scaler.fit(X_train_data)
X_train_data=scaler.transform(X_train_data)
```

```python
X_train_data=np.insert(X_train_data,4,1,axis=1)
```

```python
df=pd.DataFrame(X_train_data)
df
X_train_data.shape[1] #shape function can take axis as input.axis=0 -gives shape about rows,axis=1 -gives shape about columns.
len(X_train_data)  # gives number of rows  /  total data points
df
```

```python
def step_gradient(X_train_data,Y_train_data,learning_rate,m):
    m_slope=np.zeros(5)
    M=len(X_train_data)
    for i in range (M):
        x=X_train_data[i]   #gives row
        y=Y_train_data[i]   #gives y of row
        for j in range (X_train_data.shape[1]):
            m_slope[j]+=(-2/M)*(y-((m*x).sum()))*X_train_data[i,j]
            
    m=m-(learning_rate*m_slope) 
    return m
```

```python
def gd(X_train_data,Y_train_data,learning_rate,num_iterations):
    m=np.zeros(5)
    for k in range (num_iterations):
        m=step_gradient(X_train_data,Y_train_data,learning_rate,m)
        print(k,"cost",cost(X_train_data,Y_train_data,m))
    return m    
```

```python
def cost(X_train_data,Y_train_data,m):
    total_cost=0
    M=len(X_train_data)
    for i in range (M):
        x=X_train_data[i]
        y=Y_train_data[i]
        total_cost+=(1/M)*((y-(m*x).sum())**2)
        
    return total_cost
```

```python
def run():
    learning_rate=0.1
    num_iterations=500
    m=gd(X_train_data,Y_train_data,learning_rate,num_iterations)
    print(m)
    return m
```

```python
m=run()
```
